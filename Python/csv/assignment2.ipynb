#### 13. Ábrázoljuk a szám típusú adatok közötti összefüggést egy korrelációs mátrix segítségével!
#### Gondoskodjunk a megfelelő ábraméretről! (1.5p)

```python
#oszlopok, amikben számok vannak
numeric_columns = list(data.select_dtypes(include=[np.number]).columns)
#data[numeric_columns].head(10)

correlation_matrix = data[numeric_columns].corr()
#correlation_matrix

plt.figure(figsize=(10, 10))
plt.title("korrelációs mátrix")
plt.imshow(correlation_matrix, cmap="coolwarm")
plt.colorbar()
plt.xticks(range(len(numeric_columns)), numeric_columns, rotation=90)
plt.yticks(range(len(numeric_columns)), numeric_columns)
plt.show()
```

#### 14. Rendezzük a korrelációs mátrix értékeit a "SalePrice" oszlop szerint növekvő sorrendbe!
#### Írassuk ki azoknak az oszlopoknak a neveit, amelyek szignifikánsan hozzájárulnak a "SalePrice" alakulásához (pl. 0.3-as küszöbbel)! (1.5p)

```python
sorted_by_saleprice = correlation_matrix["SalePrice"].sort_values(ascending=True)

sorted_by_saleprice_keys = list(sorted_by_saleprice.keys())
#print(sorted_by_saleprice_keys)
#print(list(correlation_matrix.keys()))

sorted_correlation_matrix = data[sorted_by_saleprice_keys].corr()

plt.figure(figsize=(10, 10))
plt.title("korrelációs mátrix")
plt.imshow(sorted_correlation_matrix, cmap="coolwarm")
plt.colorbar()
plt.xticks(range(len(sorted_by_saleprice_keys)), sorted_by_saleprice_keys, rotation=90)
plt.yticks(range(len(sorted_by_saleprice_keys)), sorted_by_saleprice_keys)
plt.show()
sorted_by_saleprice[sorted_by_saleprice >= 0.3]
```

#### 15. Készítsünk hisztogrammot a 2006-ban eladott házak árairól! ("Yr Sold", "SalePrice" oszlopok!) 
#### A hisztogrammot osszuk összesen 15 részre az adatvizualizációhoz. 
#### Címkézzük fel helyesen a tengelyeket, adjunk címet a grafikonnak és jelenítsük meg a grafikon rácsvonalait is! (1.5p)

```python
data_2006 = data[data["Yr Sold"] == 2006]
plt.xlabel("ár")
plt.ylabel("db")
plt.hist(data_2006["SalePrice"], bins=15)
plt.title("2006-ban eladott lakások árai")

plt.grid()
plt.tight_layout()
plt.show()
```

#### 16. Készítsünk egy ábrát két subplottal a 2006-ban és 2010-ben eladott telkekről!
##### A grafikonban az x tengelyen ábrázoljuk a telkek területét ("Lot Area" oszlop), az y tengelyen pedig ábrázoljuk a telkek árát ("SalePrice" oszlop).
##### Csak azokat a házakat vegyük figyelembe, ahol a telkek területe 6000 és 10000 négyzetláb között van!
##### A grafikonon a két különböző évben eladott telkek különböző színekkel, stackplottal, rendezetten kerüljenek jelölésre!
##### Lássuk el a grafikonjainkat címmel, címkékkel, illetve gondoskodjunk a megfelelő ábraméretről is! (2p)

```python
data_big = data[data["Lot Area"] >= 6000]
data_big = data_big[data_big["Lot Area"] <= 10000]

#data_big["Lot Area"].head(20)

plt.subplot(1, 2, 1)
plt.title("2006")
plt.xlabel("terület")
plt.ylabel("ár")
data_big_2006 = data_big[data_big["Yr Sold"] == 2006]
data_big_2006_area = data_big_2006["Lot Area"].sort_values(ascending=True)
data_big_2006_price = data_big_2006["SalePrice"].sort_values(ascending=True)

plt.stackplot(data_big_2006_area, data_big_2006_price, color="red")

plt.subplot(1, 2, 2)
plt.title("2010")
plt.xlabel("terület")
plt.ylabel("ár")
data_big_2010 = data_big[data_big["Yr Sold"] == 2010]
data_big_2010_area = data_big_2010["Lot Area"].sort_values(ascending=True)
data_big_2010_price = data_big_2010["SalePrice"].sort_values(ascending=True)

plt.stackplot(data_big_2010_area, data_big_2010_price, color="orange")

plt.suptitle("2006 és 2010 lakásárak területek szerint")
plt.tight_layout()
plt.subplots_adjust(right=1.5)
plt.show()
```
